```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\Retail_sales.csv")
df_encoded=pd.get_dummies(df,columns=["Product_Category","Region"],drop_first=True)
df_encoded
```

```python
##STEP:4 BUSINESS FEATURE CREATION(REAL-WORLD)
### 1.PROFIT MARGIN

df_encoded["Profit_Margin"]=df_encoded["Profit"] / df_encoded["Sales_Amount"]
df_encoded["Profit_Margin"].fillna(0)
df_encoded
```

```python
##STEP:4 BUSINESS FEATURE CREATION(REAL-WORLD)
####2.HIFH-PROFIT LABEL(CLASSIFICATION TARGET)
df_encoded["High_Profit"]=df_encoded["Profit"].apply(lambda x:1 if x>df_encoded["Profit"].median() else 0)
df_encoded
```

```python
###### STEP:5 FEATURE SCALING(ML ESSENTIAL)

from sklearn.preprocessing import StandardScaler
#df=pd.read_csv(r"C:\Users\DELL\Downloads\Retail_sales.csv")
scaler=StandardScaler()
num_cols=["Quantity","Sales_Amount","Profit","Profit_Margin"]
df_encoded[num_cols]=scaler.fit_transform(df_encoded[num_cols])
df_encoded
```

```python
###### STEP 6:FEATURE & TARGRT SELECTION
### REGRESSION
X=df_encoded.drop(["Sales_Amount","Oder_Date","Oder_Id"],axis=1)
Y=df_encoded["Sales_Amount"]
df_encoded
```
